```python
import ipywidgets as widgets
from IPython.display import display, clear_output


# Función para mostrar el menú interactivo con botones
def display_menu():
    clear_output()  # Limpia la pantalla
    print("Selecciona una opción:")

import ipywidgets as widgets
from IPython.display import display, clear_output

# Función para mostrar el menú interactivo con botones
def display_menu():
    clear_output()  # Limpia la pantalla
    print("Selecciona una opción:")

import ipywidgets as widgets
from IPython.display import display, clear_output

# Función para mostrar el menú interactivo con botones
def display_menu():
    clear_output()  # Limpia la pantalla
    print("Selecciona una opción:")

    # Crear botones para cada opción
    button_visualizar_datos = widgets.Button(description="Visualizar datos")
    button_calcular_VPN_TIR = widgets.Button(description="Calcular VPN/TIR")
    button_graficar_resultados = widgets.Button(description="Graficar resultados")
    button_analisis = widgets.Button(description="Analisis")
    button_salir = widgets.Button(description="Salir")

    # Funciones que se ejecutan cuando se hace clic en cada botón
    def on_button_visualizar_datos_clicked(b):
        print("Visualizar datos seleccionado.")

    def on_button_calcular_VPN_TIR_clicked(b):
        print("Calcular VPN/TIR seleccionado.")

    def on_button_graficar_resultados_clicked(b):
        print("Graficar resultados seleccionado.")

    def on_button_analisis_clicked(b):
        print("Analisis resultados seleccionado.")

    def on_button_salir_clicked(b):
        clear_output()
        print("Saliendo del programa...")

    # Conectar cada botón con su respectiva función
    button_visualizar_datos.on_click(on_button_visualizar_datos_clicked)
    button_calcular_VPN_TIR.on_click(on_button_calcular_VPN_TIR_clicked)
    button_graficar_resultados.on_click(on_button_graficar_resultados_clicked)
    button_analisis.on_click(on_button_analisis_clicked)
    button_salir.on_click(on_button_salir_clicked)

    # Mostrar los botones
    display(button_visualizar_datos, button_calcular_VPN_TIR, button_graficar_resultados,button_analisis, button_salir)

# Ejecutar el menú
display_menu()
```

```python
#vincular los botones al resto del  codigo

import ipywidgets as widgets
from IPython.display import display, clear_output

import pandas as pd
import numpy as np
!pip install numpy-financial
import numpy_financial as npf
import matplotlib.pyplot as plt

#Definir df
df=pd.DataFrame()

def visualizar_datos():
    global df
    print("Visualizando datos...")
    #Importamos la información
    df=pd.read_csv('/content/Flujos_de_caja.csv', sep=';')
    df.head()
    print(df)

def calcular_vpn_tir():
    global df
    print("Calculando VPN/TIR...")
    #Calculo del VPN de los proyectos.

    # Tasa de descuento
    tasa_descuento = 0.08 # 8%

    #Flujos de caja
    columnas_flujos_de_caja = ['Inversion_Inicial','Ano_1','Ano_2','Ano_3','Ano_4','Ano_5','Ano_6','Ano_7','Ano_8','Ano_9','Ano_10']

    #Lista para almacenar los VPN por proyecto
    npvs=[]

    #Iteración por proyecto de flujos de caja
    for index in df.index:
      flujos_de_caja=df.loc[index,columnas_flujos_de_caja].tolist()

      # Cálculo del NPV
      npv = npf.npv(tasa_descuento, flujos_de_caja)
      npvs.append(npv)
      print(f"El Valor Presente Neto (VPN) por proyecto {index + 1} es: {npv:.2f}")

    #Calculo de la TIR de los proyectos.

    #Flujos de caja
    columnas_flujos_de_caja = ['Inversion_Inicial','Ano_1','Ano_2','Ano_3','Ano_4','Ano_5','Ano_6','Ano_7','Ano_8','Ano_9','Ano_10']
    #Lista para almacenar los TIR por proyecto
    tirs=[]
    #Iteración por proyecto de flujos de caja
    for index in df.index:
      flujos_de_caja=df.loc[index,columnas_flujos_de_caja].tolist()

      # Cálculo de la TIR
      tir = npf.irr(flujos_de_caja)
      tirs.append(tir)
      print(f"La Tasa Interna de Retorno (TIR) por proyecto {index + 1} es: {tir*100:.2f}%")

  #Agregar columnas VPN y TIR al Dataframe df
    df['VPN']=npvs
    df['TIR']=tirs

def graficar_resultados():
    global df
    if df.empty:
      visualizar_datos()
      calcular_vpn_tir()

    print("Graficando resultados...")
    # Grafico VPN
    df.plot(kind='bar', x='Proyecto', y='VPN', color='blue', title='Valor Presente Neto (VPN) por Proyecto')

    # Grafico TIR
    df.plot(kind='bar', x='Proyecto', y='TIR', color='green', title='Tasa Interna de Retorno (TIR) por Proyecto')
    plt.show()

    #Analisis de resultados
    #Para el analisis del VPN tomaremos los 5 proyectos mas rentables.
    # VPN
    df_sorted_vpn = df.sort_values(by='VPN', ascending=False)
    print("--- VPN ---")
    print("\nLos 5 proyectos más rentables según VPN:")
    print(df_sorted_vpn[['Proyecto', 'VPN']].head(5))
    #Para el analisis de la TIR tomaremos los 5 proyectos con ña TIR superior a la tasa descuento.
    # TIR
    df_sorted_tir = df.sort_values(by='TIR', ascending=False)
    print("--- TIR ---")
    print("\nLos 5 proyectos más rentables según TIR:")
    print(df_sorted_tir[['Proyecto', 'TIR']].head(5))

def analisis():
    print("Analisis de resultados")
    print("\nAnalizando los resultados según TIR y VPN, tenemos diferentes resultados\nel proyecto mas rentable por VPN es  Starbuck el Justo III de Troya el cual nos\nmanifiesta que tendremos un buen retorno respecto al costo de capital del \nmismo, es decir teniendo un valor adicional a la empresa que desee invertir en el.")
    print("\nPor calculo de TIR tenemos que el proyecto mas rentable seria Julia el sabio II \nya que nos está devolviendo una tasa de retorno superior a la descuento que esta en el 8%, un valor adicional del 1,8%.")
    print("\nAmbos indicadores son importantes para la toma de decisiones ya que uno nos \ninforma tasa y el otro en valor actual de los flujos, vemos que un proyecto que \nnos brinda ambas situaciones de manera positiva es Julia el sabio II, \nmostrándose como un buen proyecto de inversión.")

def salir():
    clear_output()
    print("Saliendo del programa...")

def display_menu():
    clear_output()
    print("Selecciona una opción:")

    button_visualizar_datos = widgets.Button(description="Visualizar datos")
    button_calcular_vpn_tir = widgets.Button(description="Calcular VPN/TIR")
    button_graficar_resultados = widgets.Button(description="Graficar resultados")
    button_analisis = widgets.Button(description="Analisis")
    button_salir = widgets.Button(description="Salir")

    button_visualizar_datos.on_click(lambda b: visualizar_datos())
    button_calcular_vpn_tir.on_click(lambda b: calcular_vpn_tir())
    button_graficar_resultados.on_click(lambda b: graficar_resultados())
    button_analisis.on_click(lambda b: analisis())
    button_salir.on_click(lambda b: salir())

    display(button_visualizar_datos, button_calcular_vpn_tir, button_graficar_resultados, button_analisis, button_salir)

display_menu()

```
